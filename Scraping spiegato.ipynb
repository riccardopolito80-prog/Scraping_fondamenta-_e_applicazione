```python
import requests
import json
API_KEY="api key"
#richieste ad api o siti,vediamo nel caso delle API,in altri casi di siti a seconda dello status code
#decideremo se settare nell'header il nostro account, importante in entrambi casi scegliere il
#Timeout=x secondi,in modo da non ripetere processo di scraping all'infinito, l'elaborazione
#dei dati potrà essere elaborata seguendo le indicazioni della repository sull'analisi di spotify trend
```

```python
#prendiamo come esempio l'api di un sito sul calcio , vediamo come strutturare richieste
headers={
    "x-apisports-key": API_KEY #dobbiamo sempre racchiudere la API_KEY nell'header secondo i comandi dell'API
}

#URL="https://v3.football.api-sports.io/status"
#per cambiare tipo di richiesta si cambia l'endpoint, in questo caso status
URL="https://v3.football.api-sports.io/leagues"
```

```python
richiesta=requests.request("GET",URL, headers=headers) #cosi accedi all'header con questa formula
print(richiesta.status_code)#cosi verfichiamo la nostra richiesta che abbia funzionato
print(richiesta.text)#cosi analizziamo la nostra richiesta
d=richiesta.json()#cosi la converti in dizionario json
print(json.dumps(d,indent=4)) #transformiamo in dizionari leggibili python in str
```

```python
countries=requests.request("GET", URL, headers=headers)
print(countries.json()) #come ben visto cambiando endpoint facciamo la chiamata countries
```

```python
c_dict=countries.json()['response'] #per analizzare uno specifico dato lo mettiamo nelle par quadre
print(len(c_dict)) #abbiamo c_dict come riferimento di contenitore di tutte le info
```

```python
params={
    "code":"IT"
} #per cercare nello specifico ti basi sempre sulla struttura che ti fornisce l'API
italy=requests.request("GET", URL, headers=headers,params=params)#importante sempre allegare header con API_KEY

print(json.dumps(italy.json()["response"],indent=4)) #cosi formuliamo la richiesta, abbiamo cambiato URl e ci tornerà un file json

```
